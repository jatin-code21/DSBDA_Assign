```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset("titanic")
sns.pairplot(data=titanic, hue="survived", diag_kind="hist")
plt.show()
```

```python
sns.histplot(data=titanic, x='fare', bins=30)
plt.show()
```

```python
sns.histplot(data=titanic, x='fare', log_scale=True)
plt.show()
```

```python
sns.histplot(data=titanic, x='fare', kde=True, bins = 30)
plt.show()
```

```python
sns.scatterplot(data=titanic, x='age', y='fare')
plt.show()
```

```python
sns.boxplot(data=titanic, x='class', y='fare')
plt.show()
```

```python
pivot = titanic.pivot_table(index='embark_town', columns='class', values='fare')
sns.heatmap(data=pivot, cmap='Blues')
plt.show()
```
